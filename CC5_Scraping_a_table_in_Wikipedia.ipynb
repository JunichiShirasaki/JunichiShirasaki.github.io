Step2: Identifying the URL for scraping <br>
Note: When I first tried to scrape the website, my scraping was unsuccessful. So, I am using "headers" to avoid being blocked.

```python
url = "https://en.wikipedia.org/wiki/Manufacturing"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)
tables_from_webpage = pd.read_html(url, storage_options=headers)
```

Step3: Checking the table

```python
tables_from_webpage[2].head(10)
```

Step4 : Translating the table into a variable

```python
df = tables_from_webpage[2]
df.head()
```

Step5-1: Cleaning the data (deleting unnecessary columns)

```python
df = df.drop(columns=['Rank','Year'])
df.head()
```

Step 5-2: Cleaning the data (deleting unnecessary rows)

```python
df = df[df['Country or region'].isin(['China','United States','Japan'])]
df.head()
```

Step 5-3: Cleaning the data (changing the form into "tidy")

```python
df_tidylong = df.melt(
    id_vars=['Country or region'],
    var_name= 'Measure',
    value_name= 'Value'
)
df_tidylong.head(6)
```

Step 5-4: Cleaning the data (Renaming the name of columns)

```python
df_tidylong = df_tidylong.rename(columns = {'Country or region' : 'Country'})
df_tidylong.head(6)
```

Step 6: Saving the data in the form of CSV<br>
Note: I make a chart in Vega-lite using this data.

```python
df_tidylong.to_csv('CC5_Scraping_Data_manufacturing.csv')
```
